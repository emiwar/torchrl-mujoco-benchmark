```python
import mujoco
import tqdm
import numpy as np
import mujoco._simulation_pool
```

```python
model = mujoco.MjModel.from_xml_path("models/rodent_from_dm_control.xml")
```

```python
data = mujoco.MjData(model)
```

```python
mujoco.mj_resetData(model, data)
for t in tqdm.trange(10000):
    data.ctrl = np.random.uniform(-1, 1, model.nu)
    mujoco.mj_step(model, data, 10)
```

```python
sm = mujoco._simulation_pool.SimulationPool(model, 64, 16, mujoco.mjtState.mjSTATE_FULLPHYSICS)
```

```python
sm.reset(np.ones(64, bool))
```

```python
for t in tqdm.trange(10000):
    sm.step()
```
